#ANALISIS Y LIMPIEZA

```python
#LUEGO DE LA SELECCION DE LAS COLUMNAS, COMENZAMOS LA LIMPIEZA DE DATOS DENTRO DE CADA COLUMNA.
#EN PRIMER LUGAR QUITAREMOS DE LA COLUMNA 'URL' TODAS LAS FILAS QUE CONTENGAN LAS PALABRAS 'LOCAL, COCHERA, NEGOCIO Y OFICINA'

localyoficina = nuevo_df.loc[nuevo_df['URL'].str.contains('local|oficina|cochera-en-alquiler|negocio')]
localyoficina
```

```python
#TENEMOS VARIAS FILAS QUE CONTIENEN LA PALABRA 'local|oficina|cochera-en-alquiler|negocio', PROCEDEMOS A ELIMINARLAS.

nuevo_df = nuevo_df.drop(nuevo_df[nuevo_df['URL'].str.contains('local|oficina|cochera-en-alquiler|negocio|galpon|comercio')].index)

nuevo_df.info()
```

```python
#SEGUIDAMENTE PROCEDEMOS A ELIMINAR LAS FILAS QUE SU PRECIO EN PESOS NO ALCANZAN LOS 20.000
menos20000 = nuevo_df.loc[nuevo_df['Precio ARS'] < 20000]
menos20000.info()

#ELIMINAMOS
nuevo_df = nuevo_df.drop(nuevo_df[nuevo_df['Precio ARS'] < 20000].index)

nuevo_df.info()
```

```python
#CREAMOS UN DATAFRAME DE RESPALDO CON LAS PUBLICACIONES SIN PRECIO
df_vacias = nuevo_df.loc[(nuevo_df['Precio ARS'].isnull()) & (nuevo_df['Precio USD'].isnull())]

df_vacias
```

```python
#ELIMINAMOS LAS PUBLICACIONES SIN PRECIO DEL DATAFRAME nuevo_df

nuevo_df = nuevo_df.drop(nuevo_df.loc[(nuevo_df['Precio ARS'].isnull()) & (nuevo_df['Precio USD'].isnull())].index)


nuevo_df.info()
```

```python
# CONVERTIMOS LAS COLUMNAS VALUADAS EN PESOS A SU PRECIO EN DOLAR OFICIAL DEL DIA DE LA FECHA.
dolarhoy = 365
nuevo_df['Precio USD'].fillna(df['Precio ARS'] / dolarhoy, inplace=True)
nuevo_df.info()
# TENEMOS TODAS LAS PUBLICACIONES CON SU VALOR DEL ALQUILER COTIZADO EN DOLAR BLUE.
```

```python
#EN LA COLUMNA DORMITORIOS OBSERVAMOS QUE ES DE TIPO OBJECT, Y QUE ADEMAS ENTRE SUS VALORES SE ENCUENTRA EL valor "Monoambiente"
# DEBEMOS CAMBIAR EL STRING POR  '1'.
# LUEGO CAMBIAMOS A FLOAT LOS VALORES CONTENIDOS EN LA COLUMNA
nuevo_df['Dormitorios'] = nuevo_df['Dormitorios'].replace('Monoambiente', "1")
nuevo_df['Dormitorios'] = nuevo_df['Dormitorios'].astype(float)
nuevo_df.info()
```

```python
#OBSERVAMOS LOS TIPOS DE VALORES CONTENIDOS EN LA COLUMNA "Ambientes", EN EL OBSERVAMOS QUE EXISTE UNA FILA CON EL VALOR 'ambientes ambientes'
nuevo_df["Ambientes"].value_counts()
```

```python
nuevo_df.loc[nuevo_df["Ambientes"] == 'ambientes ambientes']
```

```python
#ELIMINAMOS
nuevo_df = nuevo_df.drop(nuevo_df[nuevo_df["Ambientes"] == 'ambientes ambientes'].index)
#CONVERTIMOS LA COLUMNA A FLOAT:

nuevo_df["Ambientes"] = nuevo_df["Ambientes"].astype(float)
```

```python
nuevo_df.loc[nuevo_df["Ambientes"]>20]
```

```python
# CON RESPECTO A LA COLUMNA "Antiguedad[Años]" LA COMBINAMOS CON LA DE NOMBRE SIMILAR, RELLENANDO LOS
# VALORES VACIOS DE UNA CON LOS DE LA OTRA COLUMNA SI ES QUE TIENE VALORES
nuevo_df['Antiguedad [años]'].fillna(df['Antigüedad [años]'], inplace=True)
nuevo_df['Antiguedad [años]'].value_counts()
```

```python
#OBSERVAMOS QUE EN LUGAR DE ANTIGUEDAD TENEMOS EL AÑO DE CREACION, Y ADEMAS QUE HAY 756 FILAS CON EL VALOR "A Estrenar"
#PRIMERO REEMPLAZAMOS EL VALOR "A estrenar" por "0"
#ELIMINAMOS LA COLUMNA 'Antigüedad [años]'
nuevo_df['Antiguedad [años]'] = nuevo_df['Antiguedad [años]'].replace('A Estrenar', "0")
nuevo_df['Antiguedad [años]'] = nuevo_df['Antiguedad [años]'].astype(float)
nuevo_df= nuevo_df.drop('Antigüedad [años]', axis=1)
```

```python
nuevo_df.info()
```

```python
#COMO MUCHOS VALORES TIENEN EL AÑO DE CREACION, VAMOS A CORREGIR ESOS VALORES, FILTRANDO CON LOS REGISTREN UN VALOR MAYOR A 500
#LUEGO RESTAMOS EL VALOR DE LA FILA POR EL AÑO ACTUAL (2023)
# Filtrar los valores mayores a 500

nuevo_df['Antiguedad [años]'] = nuevo_df['Antiguedad [años]'].apply(lambda x: 2023 - x if x > 400 else x)
nuevo_df.info()
```

```python
#JUNTO LAS TABLAS BAÑOS Y TOILETTES
nuevo_df['Baños'].fillna(df['Toilettes'], inplace=True)
nuevo_df= nuevo_df.drop('Toilettes', axis=1)
```

```python
#Completo los valores vacios en baños, con 1 baño.
nuevo_df['Baños'].fillna(1, inplace=True)
nuevo_df.info()
```

```python
#REVISO LOS VALORES DE LA COLUMNA
nuevo_df['Baños'].value_counts()
```

```python
#REVISO LAS FILAS CON MAS DE 8 baños, observamos que los valores elevados corresponden a casas por ende los dejamos.
nuevo_df.loc[nuevo_df["Baños"]>8]
```

```python
nuevo_df["Estado"].value_counts()
```

```python
nuevo_df['Estado'].fillna("Bueno", inplace=True)
nuevo_df.info()
```

```python
# Convertir la columna 'Precios Usd' a enteros usando .loc
nuevo_df.loc[:, 'Precio USD'] = nuevo_df['Precio USD'].astype(int)
```

```python
#OBSERVAMOS LOS VALORES DE LA COLUMNA
unique_values = nuevo_df['Precio USD'].unique()

print(unique_values)
```

```python
# ELIMINAMOS LAS FILAS CON PRECIO USD MAYOR A 15.000...
nuevo_df = nuevo_df.drop(nuevo_df[nuevo_df['Precio USD'] > 15000].index)
```

```python
# ELIMINAMOS LAS FILAS CON Sup Cubierta MENOR A 5
nuevo_df = nuevo_df.drop(nuevo_df[nuevo_df['Sup. cubierta [m²]'] < 5].index)
```
